```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#step 1: Generate the dataset
N=500 #number of families
children_counts=np.random.choice([0,1,2,3,4],size=N,p=[0.1,0.25,0.35,0.2,0.1])
```

```python
print(children_counts)
```

```python
#step 2:compute PMF
unique,counts=np.unique(children_counts,return_counts=True)
pmf=counts/N
```

```python
print(pmf)
```

```python
#step 3:compute CDF
cdf=np.cumsum(pmf)
print(cdf)
```

```python
#step 4:plot PMF
plt.figure(figsize=(10,5))
sns.barplot(x=unique,y=pmf,color="skyblue",edgecolor="black")
plt.xlabel("Number of Children per Family")
plt.ylabel("Probability(PMF)")
plt.title("Probability Mass Function (PMF) of Family Sizes")
plt.xticks(unique)
plt.show()
```

```python
#step:5 plot CDF
plt.figure(figsize=(10,5))
# Removed the marker argument
sns.lineplot(x=unique,y=cdf,marker="o",color="red",linestyle="-")
plt.xlabel("Number of Children per Family")
plt.ylabel("cumulative Probability(CDF)")
plt.title("Cumulative Distribution Function (CDF) of Family Sizes")
plt.xticks(unique)
plt.ylim(0,1.1)
plt.show()
```
